```python
# import requried libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# set visual theme
sns.set(style='whitegrid',palette='Set2',font_scale=1.1)

# Load clean master dataset
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Banking Customer Insights/Datasets/bank_customer_master.csv')

print(df.shape)
df.head()
```

**Overview of the Data**

```python
# Summary statistics
df.describe().T

# churn percentage
churn_rate = df['Churned'].mean() * 100
print(f'Overall Churn Rate: {churn_rate: 2f}%')

# Active vs Inactive customers
print('\nActive Customer Breakdown:')
print(df['ActiveStatus'].value_counts(normalize=True)*100)
```

**Customer Demographics**

Age Distribution

```python
plt.figure(figsize=(8,5))
sns.histplot(df['Age'],kde=True,bins=25)
plt.title('Age Distribution of Customers')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

Gender vs Churn

```python
plt.figure(figsize=(6,4))
sns.countplot(data=df,x='Gender',hue='Churned')
plt.title('Churn by Gender')
plt.show()
```

Region vs Churn

```python
plt.figure(figsize=(8,5))
sns.countplot(data=df,x='Region',hue='Churned')
plt.title('Churn Rate by Region')
plt.xticks(rotation=45)
plt.show()
```

**Financial Behavior Analysis**

Credit Score vs Churn

```python
plt.figure(figsize=(8,5))
sns.boxplot(data=df,x='Churned',y='CreditScore')
plt.title('Credit Score vs Churn')
plt.show()
```

Balance vs Salary

```python
plt.figure(figsize=(8,5))
sns.scatterplot(data=df,x='Balance',y='Salary',hue='Churned',alpha=0.6)
plt.title('Balance vs Salary(Colored by Churn)')
plt.show()
```

Tenure vs Churn

```python
plt.figure(figsize=(8,5))
sns.countplot(data=df,x='Tenure',hue='Churned')
plt.title('Tenure vs Churn Rate')
plt.show()
```

**Activity Engagement Patterns**

IsActive vs Churn

```python
plt.figure(figsize=(6,4))
sns.countplot(data=df,x='ActiveStatus',hue='Churned')
plt.title('Churn by Active Status')
plt.show()
```

Complaints vs Churn

```python
plt.figure(figsize=(8,5))
sns.barplot(data=df,x='Complaints',y='Churned',estimator = lambda x:np.mean(x)*100)
plt.title('Churn Rate by Number of Complaints(%)')
plt.ylabel('Churn Rate(%)')
plt.show()
```

Products vs Churn

```python
plt.figure(figsize=(8,5))
sns.barplot(data=df,x='Products',y='Churned',estimator = lambda x:np.mean(x)*100)
plt.title('Churn Rate by Number of Products(%)')
plt.ylabel('Churn Rate(%)')
plt.show()
```

**Recency Analysis**

```python
plt.figure(figsize=(8,5))
sns.histplot(data=df,x='LastTransactionDays',hue='Churned',kde=True,bins=30)
plt.title('Recency (Last Transaction Days) vs Churn')
plt.xlabel('Days Since Last Transaction')
plt.ylabel('Count')
plt.show()
```

**Key Insights Summary**

```python
print('Key Insights')
print('1. Most churned customers have CreditScore below 650')
print('2. Customers with more than 2 complaints show >50% churn rate')
print('3. Churn rate is low for Active customers and multi-product holders')
print('4. South and West regions show higher retention rates')
```
